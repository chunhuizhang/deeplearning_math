> 因果不是观察出来的，而是干预出来的；

- 因果推理是构建世界模型的关键，不仅决定了模型能否理解世界运行的规律（rule/logic），更决定了其能否在没见到过的情况里做合理推演
- 如果 AI 没办法干预（intervention）只能观察（observe），还能学会因果吗？
    - 当前的大多数 AI 系统都是从**静态的数据**（static）中学习，不管是（text、image、还是 video）
    - 不能干预，不能动手，不能试验变量变化带来的 outcome，没有干预，因果就无法被确认；
- 人类为什么能学会因果，因为我们从小就在做实验，反复试错，哪些动作带来哪些后果；
    - 干预是人类认知中最本质的一部分；AI 却被困在“只能观察”的局限里；
- 如何在不能真正动手的前提下，让 AI 学会“像干预一样思考” =》反事实推理
    - 如果当时不是那样，而是这样，会发生什么；
    - what if
    - 在脑子里构建一个未曾发生的世界，在这个世界里进行一次虚拟干预
    - 反事实推理，是让AI在没有真实干预的条件下，也能建立因果判断的关键机制；
- 一些进展
    - 结构化 prompt：引导模型主动展开”如果不是A，会不会是B“的因果链；
    - 用反事实问答数据训练模型区分真实与假设情境；
    - 用图结构构建多模态场景：让模型识别事件之间的因果依赖；
