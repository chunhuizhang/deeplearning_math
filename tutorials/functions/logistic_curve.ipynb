logistic curve：

$$
f(x)=\frac{L}{1+e^{-k(x-x_0)}}
$$

descending logistic curve:

$$
f(x)=\frac{L}{1+e^{k(x-x_0)}}
$$

sigmoid function (logistic curve when $L=1,k=1,x_0=0$):

$$
\sigma(x)=\frac{1}{1+e^{-x}}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义参数
L = 1        # 曲线的最大值
k = 1        # 曲线的陡峭程度
x0 = 0       # 曲线的中心点

# 生成x数据
x = np.linspace(-10, 10, 400)

# 定义下降的逻辑斯蒂曲线函数
def descending_logistic(x, L, k, x0):
    return L / (1 + np.exp(k * (x - x0)))

# 计算y数据
y = descending_logistic(x, L, k, x0)

# 绘制曲线
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='descending logistic curve')

# 添加标题和标签
plt.title('descending logistic curve')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)

# 显示图形
plt.show()
```
