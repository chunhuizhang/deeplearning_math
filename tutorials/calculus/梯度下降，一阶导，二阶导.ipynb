## 导数为0的点

- 函数一阶导数 $f'(x)=0$ 的点，是驻点；
- 如果一个点是驻点，并且在这一点函数取得局部最大值或局部最小值，那么这个点被称为极值点。
    - 判断极值的一个方法是通过二阶导数测试（如果 $f''(x)\gt 0$，那么 $f(x)$ 在 $x$ 处有极小值，
    - 如果 $f''(x)\lt 0$，则其在 $x$ 处有极大值；

## $x^2$

```python
xs = np.arange(-5, 5, 0.1)
plt.plot(xs, xs**2)
```

```python
f = x**2
```

```python
f.diff(x)
```

```python
sp.solve(f.diff(x))
```

```python
# possive => minimum
hessian(f, [x,])
```

```python
f = -x**2
print(f.diff(x))
print(sp.solve(f.diff(x)))
# negative => maximum
hessian(f, [x, ])
```

## $x^3$

```python
xs = np.arange(-5, 5, 0.1)
plt.plot(xs, xs**3)
```

```python
f = x**3
print(f.diff(x))
print(sp.solve(f.diff(x)))
# 在 x = 0 处，hessian = 0，neither
hessian(f, [x, ])
```

## 梯度下降

$$
J(x)=x^4-6x^3+8x^2
$$

```python
xs = np.arange(-1.5, 5, .1)
plt.plot(xs, xs**4 - 6*xs**3 + 8*xs**2)
```

```python
J = x**4-6*x**3+8*x**2
```

```python
J.diff(x)
```

```python
sp.solve(J.diff(x))
```

```python
sp.solve(J.diff(x))[0].evalf(), sp.solve(J.diff(x))[1].evalf(), sp.solve(J.diff(x))[2].evalf()
```
