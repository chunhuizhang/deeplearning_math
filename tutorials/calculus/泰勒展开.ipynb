## bascis


- 函数的泰勒展开是一种近似策略
    - 在 $x=a$ 处展开
    $$
    f(x)=f(a)+f'(a)(x-a)+\frac{f''(a)}{2!}(x-a)^2+\frac{f'''(a)}{3!}(x-a)^2+\cdots
    $$
    - 其中一阶泰勒展开的形式，一定是线性的，即沿着 $x=a$ 处的切线（斜率为 $f'(a)$）进行一阶线性逼近
- 对于二元函数 $f(x,y)$

    $$
    \begin{split}
    f(x,y)&=f(a,b)+f_x(a, b)(x-a)+f_y(a,b)(y-b)\\
    &+\frac1{2!}\left((x-a)^2f_{xx}(a,b)+2(x-a)(y-b)f_{xy}(a, b)+(y-b)^2f_{yy}(a,b)\right)\\
    &+\cdots
    \end{split}
    $$
    - 其一阶形式为两个切平面的线性求和；

### examples

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.dpi'] = 120
```

```python
def f(x):
    return 2*x**3 + 3*x**2 + 4*x + 1 
```

```python
x = np.arange(-5, 5, .01)
plt.plot(x, f(x))
```

```python
f_1_prime = lambda x: 6*x**2 + 6*x + 4
f_2_prime = lambda x: 12*x + 6
f_3_prime = lambda x: 12
```

```python
f_1 = lambda x: f(0) + f_1_prime(0)*x
f_2 = lambda x: f(0) + f_1_prime(0)*x + f_2_prime(0)/2 * x**2
f_3 = lambda x: f(0) + f_1_prime(0)*x + f_2_prime(0)/2 * x**2 + f_3_prime(0)/6* x**3
```

```python
plt.plot(x, f(x), label='f(x)')
plt.plot(x, f_1(x), label='f_1')
plt.plot(x, f_2(x), label='f_2')
# plt.plot(x, f_3(x), label='f_3')
plt.legend()
```
