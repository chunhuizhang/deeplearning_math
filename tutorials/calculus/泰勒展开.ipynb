- 一阶 & 二阶泰勒展开
    - 一阶展开（即线性近似）
    - 二阶展开加入了曲率信息
        - 几何上表示的是凹凸性

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.dpi'] = 120
```

## bascis

> 任意连续函数都可以被分解为泰勒级数（一种幂级数 $(x-x_0)^k$，幂级数的求导最为方便）；


- 函数的泰勒展开是一种近似策略
    - 在 $x=a$ 处展开
    $$
    f(x)=f(a)+f'(a)(x-a)+\frac{f''(a)}{2!}(x-a)^2+\frac{f'''(a)}{3!}(x-a)^2+\cdots
    $$
    - 其中一阶泰勒展开的形式，一定是线性的，即沿着 $x=a$ 处的切线（斜率为 $f'(a)$）进行一阶线性逼近

    $$
    f(x)\approx f(x_0)+f'(x_0)(x-x_0)
    $$
    
- 对于二元函数 $f(x,y)$

    $$
    \begin{split}
    f(x,y)&=f(a,b)+f_x(a, b)(x-a)+f_y(a,b)(y-b)\\
    &+\frac1{2!}\left((x-a)^2f_{xx}(a,b)+2(x-a)(y-b)f_{xy}(a, b)+(y-b)^2f_{yy}(a,b)\right)\\
    &+\cdots
    \end{split}
    $$
    - 其一阶形式为两个切平面的线性求和；

## 几何性质

$$
f(x)=f(x_0)+f'(x_0)(x-x_0)+\frac{f''(x_0)}{2!}(x-x_0)^2+\cdots
$$

- 0阶近似：$f(x)=f(x_0)$
- 1阶近似：$f(x)=f(x_0)+f'(x_0)(x-x_0)$（与原函数在$x_0$处一阶导数相同）
- 2阶近似：$f(x)=f(x_0)+\frac{f''(x_0)}{2!}(x-x_0)^2$（与原函数在$x_0$处一阶导数，二阶导数相同）

### 二阶导数凹凸性的刻画

- $f(x)=x^3-3x^2+2x$
    - $f'(x)=3x^2-6x$
    - $f''(x)=6x-6$
    - $f''(x)\gt 0$ 时（$x\gt 1$），原函数是凸的，$f''(x)\lt 0$时（$x\lt 1$），原函数是凹的；

```python
x = np.linspace(-1, 4, 400)
f = x**3 - 3*x**2 + 2*x
f_prime = 3*x**2 - 6*x + 2
f_double_prime = 6*x - 6

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, f, label='f(x) = $x^3 - 3x^2 + 2x$', linewidth=2)
plt.plot(x, f_prime, label="f'(x) = $3x^2 - 6x + 2$", linestyle='--')
plt.plot(x, f_double_prime, label="f''(x) = $6x - 6$", linestyle='-.')
plt.title('Function and Its First and Second Derivatives')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.grid(True)
plt.legend()
```

```python
def f(x):
    return 2*x**3 + 3*x**2 + 4*x + 1 
```

```python
x = np.arange(-5, 5, .01)
plt.plot(x, f(x))
```

```python
f_1_prime = lambda x: 6*x**2 + 6*x + 4
f_2_prime = lambda x: 12*x + 6
f_3_prime = lambda x: 12
```

```python
f_1 = lambda x: f(0) + f_1_prime(0)*x
f_2 = lambda x: f(0) + f_1_prime(0)*x + f_2_prime(0)/2 * x**2
f_3 = lambda x: f(0) + f_1_prime(0)*x + f_2_prime(0)/2 * x**2 + f_3_prime(0)/6* x**3
```

```python
plt.plot(x, f(x), label='f(x)')
plt.plot(x, f_1(x), label='f_1')
plt.plot(x, f_2(x), label='f_2')
# plt.plot(x, f_3(x), label='f_3')
plt.legend()
```
