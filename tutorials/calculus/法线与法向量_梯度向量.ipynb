- 法线（normal line）与法向量（normal vector）
    - 在一元函数的上下文中，法向量可以被看作是垂直于函数图像的直线的方向向量，对于 $y=ax+b$ 其斜率为 $a$，则垂直于这条直线的直线的斜率为 $-1/a$（相乘为-1）
    - 一元函数的斜率，对边/邻边

- 考虑一个二元函数 $x^4+xy+y^2=19$，在 $P(2,-3)$ 点的切平面（tangent plane）
    - $f(x,y)=x^4+xy+y^2-19$
    - $\nabla_xf(x,y)=4x^3+y, \nabla_yf(x,y)=x+2y$
    - 在点 $P(2,-3)$ 处的梯度向量为 $(29,-4)$
    - 则切平面被定义为 $29(x-2)-4(y+3)=0$

### 法向量和梯度向量

- 法向量定义在平面上
    - $2x+3y-4z=7$, $\mathbf n=(2,3,-4)$
- 梯度向量定义在函数上
    - 平面不是函数；
    - 这个梯度向量在三维空间中垂直于曲面在点 $(x,y,f(x,y))$ 处的切平面。

```python
import matplotlib.pyplot as plt
import numpy as np

# 定义一元线性函数 y = ax + b，这里选择 a = 2, b = 1
a = 2
b = 1
def linear_function(x):
    return a * x + b

# 定义一元线性函数的法线方向函数
def perpendicular_line(x, x0, y0):
    # 斜率为 -1/a
    perpendicular_slope = -1./a
    return perpendicular_slope * (x - x0) + y0

# 创建 x 的值
x = np.linspace(-10, 10, 400)
y = linear_function(x)

# 选择一个点在直线上 (比如 x = 5)，计算该点的 y 值和法线方向
x0 = 5
y0 = linear_function(x0)
y_perpendicular = perpendicular_line(x, x0, y0)

# 绘制直线和法线
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Linear Function y = 2x + 1')
plt.plot(x, y_perpendicular, label='Perpendicular at x = 5')
plt.scatter([x0], [y0], color='red') # 标记选择的点
plt.axis('equal')

# 设置图表的标题和图例
plt.legend()
```
