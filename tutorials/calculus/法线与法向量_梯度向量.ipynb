```python
import matplotlib.pyplot as plt
import numpy as np

#  y = ax + b a = 2, b = 1
a = 2
b = 1
def linear_function(x):
    return a * x + b

# 
def perpendicular_line(x, x0, y0):
    #  -1/a
    perpendicular_slope = -1./a
    return perpendicular_slope * (x - x0) + y0

#  x 
x = np.linspace(-10, 10, 400)
y = linear_function(x)

#  ( x = 5) y 
x0 = 5
y0 = linear_function(x0)
y_perpendicular = perpendicular_line(x, x0, y0)

# 
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Linear Function y = 2x + 1')
plt.plot(x, y_perpendicular, label='Perpendicular at x = 5')
plt.scatter([x0], [y0], color='red') # 
plt.axis('equal')

# 
plt.legend()
```
