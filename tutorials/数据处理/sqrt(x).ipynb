```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from 0 to 100
x = np.linspace(0, 100, 500)

# Calculate y values for each function
y1 = x
y2 = np.sqrt(x) * 10

# Create the plot
plt.plot(x, y1, label='y = x')
plt.plot(x, y2, label='y = sqrt(x) * 10')
plt.axhline(y=60, color='grey', linestyle='--')
plt.axvline(x=36, color='grey', linestyle='--')
# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x and y = sqrt(x) * 10')

plt.legend()
```
