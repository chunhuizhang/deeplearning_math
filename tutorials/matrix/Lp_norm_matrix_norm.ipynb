## Lp norm

$$
\|x\|_p=\left(\sum_i^nx_i^p\right)^{1/p}
$$

- $L_p$ norm
    - $L_1$: $\|x\|_1=\sum_i|x_i|$
    - $L_2$: $\|x\|_2=(\sum_ix_i^2)^{1/2}$
    - $L_{\infty}$: $\|x\|_\infty=\max_i|x_i|$


### 可视化、几何分析

$$
\|x\|_p=1
$$

- 表示的是一个集合

## matrix norm

- 从向量的 $L_p$ norm 诱导而来（induced norm）
    - biggest $L_p$ norm
    - $\|AB\|_p\leq \|A\|_p\|B\|_p$
    - $\|A\|_1$ 列和最大；
    - $\|A\|_\infty$ 行和最大；
    - $\|A\|_2$ 谱范数；
    - $\|A\|_{\rm F}$ , Frobenius norm.
$$
\begin{split}
&\|A\|_p=\max_{\|x\|_p}\frac{\|Ax\|_p}{\|x\|_p}\\
&\|A\|_1=\max_{1 \leq j \leq n} \sum_{i=1}^m | a_{ij}|\\
&\|A\|_\infty = \max_{1 \leq i \leq m} \sum _{j=1}^n | a_{ij} |\\
&\|A\| _2 = \sigma_{\max}(A) \leq \|A\|_{\rm F} = \left(\sum_{i=1}^m \sum_{j=1}^n |a_{ij}|^2\right)^{\frac{1}{2}} = \left(\sum_{k=1}^{\min(m,n)} \sigma_{k}^2\right)^{\frac{1}{2}}
\end{split}
$$
