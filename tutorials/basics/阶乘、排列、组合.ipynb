## 阶乘

$$
\begin{split}
&\Gamma(x)=\int_0^\infty u^{x-1}e^{-u}du\\
&\Gamma(n+1)=n!\\
&\Gamma(n)=(n-1)!
\end{split}
$$

- $\Gamma(\cdot)$ 在负整数点上为 `inf`

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma, factorial
x = np.linspace(-3.5, 5.5, 2251)
y = gamma(x)
plt.plot(x, y, label='$\Gamma(x)$')
k = np.arange(1, 7)
plt.plot(k, factorial(k-1), 'k*', alpha=.6)
plt.ylim(-10, 10)
plt.xlim(-3.5, 5.5)
```

## 排列（permutation）

$$
P(n, r)=A_{n}^{r}=\frac{n!}{(n-r)!}
$$

- 这里可以得到一个 $0!=1$ 的定义/证明
    - $A_n^{n}=\frac{n!}{(n-n)!}=\frac{n!}{0!}=n!$

## 组合（combination）

$$
\begin{split}
C(n,r)&=\binom{n}{r}=\binom{n}{n-r}=\frac{n!}{r!(n-r)!}\\
&=\frac{A_n^r}{r!}
\end{split}
$$

$$
\binom n0 + \binom n1+ \cdots + \binom n{n-1} + \binom nn=2^n
$$

- 杨辉三角的每一行为二项式系数，行和为2的幂

## 总结

- Permutations (order matters):	顺序有关
- Combination (order doesn't matter): 顺序无关
