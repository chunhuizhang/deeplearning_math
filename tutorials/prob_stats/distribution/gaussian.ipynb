```python
import numpy as np
from matplotlib import pyplot as plt
```

## isotropic

- https://math.stackexchange.com/questions/1991961/gaussian-distribution-is-isotropic

```python
pts = np.random.multivariate_normal([0, 0], 
                                    [[5, 0], [0, 5]], 
                                    size=10000, check_valid='warn')

plt.scatter(pts[:, 0], pts[:, 1], s=1)
plt.xlim((-10,10))
plt.ylim((-10,10))
```

```python
pts = np.random.multivariate_normal([0, 0], 
                                    [[1, 0], [0, 5]], 
                                    size=10000, check_valid='warn')

plt.scatter(pts[:, 0], pts[:, 1], s=1)
plt.xlim((-10,10))
plt.ylim((-10,10))
```

```python
pts = np.random.multivariate_normal([0, 0], 
                                    [[1,15], [15, 31]], 
                                    size=10000, check_valid='warn')

plt.scatter(pts[:, 0], pts[:, 1], s=1)
plt.xlim((-20,20))
plt.ylim((-20,20))
```
