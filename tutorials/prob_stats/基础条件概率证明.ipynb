- 条件概率中依然可以有联合概率
    - $P(A,B|C)$：A和B的 joint given C
- 全概率公式依然是从条件概率中来
    - $P(X)=\sum_iP(X_i|Y_i)P(Y_i)$

## $P(A|B,C)=\frac{P(A,B|C)}{P(B|C)}$

$$
\begin{split}
P(A|B,C)=&\frac{P(A,B,C)}{P(B,C)}\\
=&\frac{\frac{P(A,B,C)}{P(C)}}{\frac{P(B,C)}{P(C)}}\\
=&\frac{P(A,B|C)}{P(B|C)}
\end{split}
$$

$$
\begin{split}
&P(A,B|C)=P(B|C)P(A|B,C)\\
&P(A,B|C)=P(A|C)P(B|A,C)
\end{split}
$$

## 条件独立

- https://en.wikipedia.org/wiki/Conditional_independence

$$
P(A\mid B,C) = P(A \mid C)
$$

- $A$ and $B$ are said to be conditionally independent given $C$, written symbolically as: $\displaystyle (A\perp \!\!\!\perp B\mid C)$. 

$$
\begin{split}
P(A\mid B,C)&=P(A\mid C)\\
\Updownarrow\\
\frac{P(A,B,C)}{P(B, C)}&=\frac{P(A,C)}{P(C)}\\
\Updownarrow\\
P(A,B,C)&=\frac{P(B,C)P(A,C)}{P(C)}\\
\Updownarrow\\
\frac{P(A,B,C)}{P(C)}&=\frac{P(B,C)}{P(C)}\frac{P(A,C)}{P(C)}\\
\Updownarrow\\
P(A,B|C)&=P(A|C)P(B|C)
\end{split}
$$

## 一个 mdp 应用

$$
P(s'|a)=\sum_sP(s|a)P(s'|s,a)
$$

证明如下：

$$
\begin{split}
P(s,s'|a)&=P(s|a)P(s'|s,a)\\
P(s'|a)&=\sum_sP(s,s'|a)\\
P(s'|a)&=\sum_sP(s|a)P(s'|s,a)
\end{split}
$$

- 其实上面的 $a$ 可以替换为更多的变量 $(b,a,z)$

$$
P(s'|b,a,z)=\sum_sP(s|b,a,z)P(s'|s,b,a,z)
$$
